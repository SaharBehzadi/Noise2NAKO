## Example for Forest-Guided Clustering on regression Random Forest model

```python
data_path = '../../Data/KORA_S3_S4/KORA_Noise_CleanData_MiniModel.csv'
data = pd.read_csv(data_path)
# rf = joblib.load(open('./data/random_forest_boston.joblib', 'rb'))
```

```python
X = data.drop(['bp_syst'], axis = 1)
y = data['bp_syst']
```

```python
rf = RandomForestRegressor(n_estimators= 10, max_depth= 10, random_state= 0, max_features='sqrt').fit(X, y)
```

```python
optimal_k = forest_guided_clustering(output='KORA_S3_S4_FGC_result.png', model=rf, data=data, target_column='bp_syst',  
                                     max_K = 6, thr_pvalue = 0.001, 
                                     bootstraps = 100, max_iter_clustering = 200, discart_value = 0.6, number_of_clusters = None, random_state = 10)
```
